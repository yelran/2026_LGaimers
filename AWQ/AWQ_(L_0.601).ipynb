```python
import torch, random, os
import numpy as np

print("torch:", torch.__version__)
print("cuda:", torch.cuda.is_available(), "| gpu:", torch.cuda.get_device_name(0))

random.seed(0)
np.random.seed(0)
torch.manual_seed(0)
torch.cuda.manual_seed_all(0)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

MODEL_ID = "LGAI-EXAONE/EXAONE-4.0-1.2B"

tokenizer = AutoTokenizer.from_pretrained(MODEL_ID, trust_remote_code=True)

model = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float16,  
    trust_remote_code=True,
)
model.eval().to("cuda")

print("Loaded:", next(model.parameters()).device, next(model.parameters()).dtype)
```

```python
from datasets import load_dataset

DATASET_ID = "LGAI-EXAONE/MANTA-1M"
DATASET_SPLIT = "train"

NUM_CALIBRATION_SAMPLES = 64
MAX_SEQUENCE_LENGTH = 256

ds = load_dataset(DATASET_ID, split=f"{DATASET_SPLIT}[:{NUM_CALIBRATION_SAMPLES}]")

def preprocess(ex):
    return {
        "text": tokenizer.apply_chat_template(
            ex["conversations"],
            add_generation_prompt=True,
            tokenize=False
        )
    }

ds = ds.map(preprocess)

# text만 남기기
ds = ds.remove_columns([c for c in ds.column_names if c != "text"])

print("calib samples:", len(ds))
print(ds[0]["text"][:200])
```

```python
from llmcompressor import oneshot
from llmcompressor.modifiers.awq import AWQModifier

# AWQ 4bit group quant (일반적으로 많이 쓰는 설정)
AWQ_CONFIG_GROUPS = {
    "group_0": {
        "targets": ["Linear"],
        "weights": {
            "num_bits": 4,
            "type": "int",
            "symmetric": False,
            "strategy": "group",
            "group_size": 128
        }
    }
}

def run_awq(recipe):
    oneshot(
        model=model,
        dataset=ds,
        recipe=recipe,
        max_seq_length=MAX_SEQUENCE_LENGTH,
        num_calibration_samples=NUM_CALIBRATION_SAMPLES,
    )

# 1) 자동 매핑 시도
recipe_auto = [
    AWQModifier(
        mappings=None,                 
        ignore=["lm_head"],
        config_groups=AWQ_CONFIG_GROUPS,
        offload_device=None,
    )
]

try:
    print("[AWQ] Trying auto mappings...")
    run_awq(recipe_auto)
    print("[AWQ] Auto-mapping AWQ done ")
except Exception as e:
    print("[AWQ] Auto mappings failed -> fallback mappings used")
    print("Reason:", repr(e))

    # 2) fallback: 넓게 커버하는 regex 매핑(EXAONE 구조가 llama류와 유사하다는 가정 하에 범용 처리)
    fallback_mappings = [
        {
            "smooth_layer": "re:.*(input_layernorm|self_attn_layer_norm|attention.*layer_norm|self_attn.*layer_norm).*",
            "balance_layers": ["re:.*q_proj", "re:.*k_proj", "re:.*v_proj"]
        },
        {
            "smooth_layer": "re:.*(post_attention_layernorm|final_layer_norm|norm).*",
            "balance_layers": ["re:.*(fc1|up_proj|gate_proj).*"]
        },
    ]

    recipe_fallback = [
        AWQModifier(
            mappings=fallback_mappings,
            ignore=["lm_head"],
            config_groups=AWQ_CONFIG_GROUPS,
            offload_device=None,
        )
    ]

    print("[AWQ] Trying fallback mappings...")
    run_awq(recipe_fallback)
    print("[AWQ] Fallback-mapping AWQ done ")
```

```python
import os

OUT_DIR = "model"
os.makedirs(OUT_DIR, exist_ok=True)

model.save_pretrained(OUT_DIR, save_compressed=True)
tokenizer.save_pretrained(OUT_DIR)

print("Saved:", OUT_DIR)
print("Files:", os.listdir(OUT_DIR))
```
